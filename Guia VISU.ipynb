```python
import pandas as pd

estudiantesMat = pd.read_csv('resources/student-mat.csv')
estudiantesPor = pd.read_csv('resources/student-por.csv')

estudiantesMat.head()

estudiantesPor.head()
```

# Pregunta 1

```python
# D )Existen mas alumnos de Portugués en la escuela Mousinho da Silveira que en la Gabriel Pereira

print(estudiantesPor.columns)
```

```python
estudiantesPor['cantAlumnos'] = 1
estudiantesPorEscuela = estudiantesPor.groupby('school').count()
estudiantesPorEscuela.sort_values('cantAlumnos', ascending=False)
```

```python
# B) Los alumnos cuyas madres están en la casa tienen mejor desempeño en Matemática que resto de los alumnos 

estudiantes = pd.concat([estudiantesMat, estudiantesPor])
```

```python
estudiantes.head()
```

```python
print(estudiantes['Fjob'].unique())
```

```python
mamasEnCasa = estudiantesMat.groupby('Mjob').mean()
mamasEnCasa.sort_values('G3',ascending=False)
```

```python
# A) a)  El nivel de educación de la madre tiene mas influencia sobre las calificaciones de los hijos, que el nivel de educación del padre.

eduacionMamas = estudiantes.groupby('Medu').mean()
eduacionMamas.sort_values('G3',ascending=False)
```

```python
eduacionPapas= estudiantes.groupby('Fedu').mean()
eduacionPapas.sort_values('G3',ascending=False)
```

```python
# El consumo de alcohol durante los fines de semana tiene mas influencia en la salud de los alumnos que en el tiempo de estudio.

estudiantesSalud = estudiantes.groupby('health').mean()
estudiantesSalud.head()
```

# PREGUNTA 2

```python
ufo_data = pd.read_csv('resources/scrubbed.csv', low_memory=False)
ufo_data['datetime'] = pd.to_datetime(ufo_data['datetime'], errors='coerce')
ufo_data['duration (seconds)'] = pd.to_numeric(ufo_data['duration (seconds)'], errors='coerce')
```

```python
ufo_data.columns
```

```python
# A) CANADA 2010
canada = ufo_data[ufo_data['country'] == 'ca']
canada2010 = canada[pd.DatetimeIndex(canada['datetime']).year == 2010]
canada2010.describe()
```

```python
# b)   La forma con mayor cantidad de registros en Los Angeles, CA es la forma de triangulo. 

losAngeles = ufo_data[ufo_data['city'].str.contains('ngeles')]
losAngeles = losAngeles[losAngeles['state'] == 'ca']
losAngeles['totalPorTipo'] = 1
losAngelesByShape = losAngeles.groupby('shape').sum()
losAngelesByShape = losAngelesByShape.sort_values('totalPorTipo',ascending=False)
losAngelesByShape.head()
```

```python
#c)  Existen 963 ufos reportados con forma de chevron 
ufo_data['totalPorTipo'] = 1
ufoPorTipo =ufo_data.groupby('shape').sum()
ufoPorTipo.sort_values('totalPorTipo',ascending=False)
```
